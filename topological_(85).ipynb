```python
from scipy.spatial.distance import cdist
s = cdist(u3, u3, 'cosine')
s
```

```python
X1=pd.DataFrame(s)
X1
```

```python
# taking average of the hamming distance matrix
y = np.average(X1)
y
```

```python
adjmat_df=X1.copy()
```

```python
# setting threshold value as mean of hamming matrix
adjmat_df[adjmat_df < y]=0
adjmat_df[adjmat_df >= y]=1
```

```python
# adjacency matrix
adjmat_df
```

```python
classList=list(df1['Attrition'])
```

```python
# Create networkx graph object
graphNw = nx.from_pandas_adjacency(adjmat_df,create_using = nx.Graph())

#print(nx.info(graphNw))
```

```python
edge_connnected=list(graphNw.edges())
```

```python
#######Code for Node self  loop in case, if any node is absent in the edge file 
noOfNodes = graphNw.number_of_nodes()
edge_connnected=list(graphNw.edges())
edge_list= np.array(edge_connnected)
srcNodeList=list(edge_list[:,0])
targetNodeList=list(edge_list[:,1])
```

```python
for i in range(noOfNodes):
    if (not i in srcNodeList) and (not i in targetNodeList):
        graphNw.add_edge(i,i)
        edge_connnected.append([i,i])
        adjmat_df.iat[i,i]=1.15
        adjmat_df.iat[i,i]=1
```

```python
edge_connnected=list(graphNw.edges())
edgesDF = pd.DataFrame(edge_connnected)
#edgesDF.to_csv('graphNwEdges.csv', encoding='utf-8',index=False, header=False)
```

```python
diGraph = nx.DiGraph()
diGraph.add_edges_from(edge_connnected)
#diGraph = nx.from_pandas_edgelist(edge_connnected, source='u1', target='u2', create_using=nx.DiGraph())
#print(nx.info(diGraph))                           
```

```python
in_degreeDict = dict(diGraph.in_degree())

in_degreeDict = collections.OrderedDict(sorted(in_degreeDict.items()))
inDegreeList=list(in_degreeDict.values())
```

```python
out_degreeDict = dict(diGraph.out_degree())
out_degreeDict=collections.OrderedDict(sorted(out_degreeDict.items()))
outDegreeList=list(out_degreeDict.values())
```

```python
hubDict,authDict = nx.hits(graphNw)
hubList=list(hubDict.values())
authList=list(authDict.values())
```

```python
from networkx.algorithms import community
communities = community.greedy_modularity_communities(graphNw)
```

```python
modularity_dict = {} # Create a blank dictionary
for i,c in enumerate(communities): # Loop through the list of communities, keeping track of the number for the community
    for name in c: # Loop through each person in a community
        modularity_dict[int(name)] = i
modularityDict=collections.OrderedDict(sorted(modularity_dict.items()))
modularityList=list(modularityDict.values())
```

```python
PageRankDict=nx.pagerank(graphNw, alpha=0.85)
PageRankList=list(PageRankDict.values())
```

```python
triangleList=[]
for node in graphNw.nodes():
    triangleList.append(nx.triangles(graphNw, node))
#triangleList
```

```python
closenessDict=nx.closeness_centrality(graphNw)
closenessList=list(closenessDict.values())
```

```python
betweennessCentralityDict = nx.betweenness_centrality(graphNw)
betweennessCentralityList=list(betweennessCentralityDict.values())
```

```python
harmonicCentralityDict=nx.harmonic_centrality(graphNw)
harmonicCentralityList=list(harmonicCentralityDict.values())
```

```python
eigenvectorDict = nx.eigenvector_centrality(graphNw) 
eigenvectorList=list(eigenvectorDict.values())
```

```python
graphFeatures_dict = {'InDegree':inDegreeList,'OutDegree':outDegreeList,'Degree':degreesList,'WeightedInDegree':inDegreeList,
              'ClosenessCentrality':closenessList,'HarmonicCentrality':harmonicCentralityList,
              'BetweennessCentrality' :betweennessCentralityList,'Authority':authList,'hub':hubList,'Modularity':modularityList,
              'PageRank':PageRankList,'Triangles':triangleList,'EigenvectorCentrality':eigenvectorList,'Class':classList}
graphFeaturesDf = pd.DataFrame.from_dict(graphFeatures_dict)
graphFeaturesDf
```
